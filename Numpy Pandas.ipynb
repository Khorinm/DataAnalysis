__№1__.(Completed)  Написать функцию, которая берет на вход прямоугольную матрицу $A$ размера m x n и превращает каждый ее столбец в дискретное распределение с помощью применения функции _softmax_:

$\text{softmax}(a_j) = \frac {\exp (a_j)}{\sum_i \exp(a_i)}$, где a --- столбец матрицы $A$.

Таким образом, в матрице, возвращенной функцией, сумма каждого столбца матрицы должна быть равна 1.

```python
m = np.random.randint(2, 6)
n = np.random.randint(2, 6)
A = np.random.randint(0, 3,(m, n))

def compute_softmax(A):
    A = np.exp(A) / np.sum(np.exp(A), axis = 0)
    return A

print('Input:\n', A)
A = compute_softmax(A)
print('Output:\n', A)

#Для проверки
print('Сумма по столбцам:', np.sum(A, axis = 0))    
```

```python
#Пример
```

```python
n = np.random.randint(2, 8)
A = np.array(np.random.randint(1,21,(1,n)), dtype = int)

def transform_matrix(A):
    A = A.T * np.eye(n, dtype = int)
    A_changed = np.hstack((np.zeros((n,1), dtype = int), A))
    A = np.vstack((A_changed, np.zeros((1,n+1), dtype = int)))
    return A

print('Source:\n', A)
print('Transformed matrix:\n', transform_matrix(A))
```

__№7__.(Completed) Дан массив X (размерность неизвестна). Заменить все элементы, кратные 3, которые больше 0 и меньше 100, на среднее значение по массиву.

```python
#Пример
```

```python
import pandas as pd
n = np.random.randint(1,6)
m = np.random.randint(1,6)
X = np.random.randint(0,50,(n,m))

def ChangeMatrix(X):
    X[(X % 3 == 0) & np.logical_and( X > 0, X < 100)] = np.mean(X)
    return X

print('Input:\n', X)
print('Output:\n', ChangeMatrix(X))
```

__№8__.(Completed) Напишите функцию, которая берет на вход датафрейм pandas со считанными данными и возвращает ответ на следующий вопрос (строка).

Найдите страну, у которой наибольшее ВВП среди стран площадью меньше 200 или население меньше 10000 (или — не исключающее).

```python
df = pd.read_csv('countries.csv', sep=';')

def find_country(df):
    sample = df.loc[(df["Area(sq km)"] < 200) | (df["Population"] < 10000)]
    maxGDP = np.max(sample['GDP'])
    country = sample.Country[sample.GDP == maxGDP]
    return str(country).split('\n')[0].split('    ')[1]

print('Страна, у которой наибольшее ВВП среди стран площадью меньше 200 или населением меньше 10000:', find_country(df))
```

__№10__.(Completed)  Напишите функцию, которая берет на вход датафрейм pandas и число и возвращает ответ на вопрос (строка):

В какой стране ВВП (GDP) наиболее близок к указанной величине?

```python
df = pd.read_csv('countries.csv', sep=';')

def compare_GDP(df, gdp):
    df.insert(loc = 10, column = "GDP_diff", value = np.abs(df.GDP - gdp))
    return str(df.Country[df.GDP_diff == np.min(df.GDP_diff)]).split('\n')[0].split('    ')[1]

print("Страна, в которой ВВП наиболее близок к указанной величине:", compare_GDP(df, 39396030702))
```

Напишите функцию, которая берет на вход датафрейм pandas со считанными данными и возвращает датафрейм с двумя колонками: буква алфавита и значение ВВП, усреднённое по всем странам, начинающимся с данной буквы.

Для решения задачи может понадобиться функция pandas.groupby, которую вы не успели пройти на семинаре. Поэтому задача и дополнительная.

```python
df = pd.read_csv('countries.csv', sep=';')

def compute_means_by_letters(df):
    condition = [df.loc[df['Country'].str.startswith(chr(x))].mean()['GDP'] for x in range(ord('A'), ord('Z') + 1)]
    new_df = pd.DataFrame({'Letter' : [chr(x) for x in range(ord('A'), ord('Z') + 1)], 'GDP' : condition})
    return new_df.dropna() #Теперь обрабатывает NaN

compute_means_by_letters(df)
```
