Преобразуем все номинальные признаки в несколько признаков с бинарным значением. Поскольку часть номинальных признаков бинарные, перекодируем их значения в 0 и 1 через LabelEncoder, а для номинальных признаков со значением большим, чем 2 будем применять one-hot encoding.

```python
from sklearn.preprocessing import LabelEncoder

def encode_nominals(column_names, df):

    """column_names - names of nominal features in datafraem, df - dataframe"""
    
    print("Nominal features encoding started...")
    encoder = LabelEncoder()
    dicts = {} # Keeps initial values of binary features
    for name in column_names:
        # Get column index
        index = df.columns.get_loc(name)
        encoder.fit(df[df.columns[index]].drop_duplicates())
        # Watch labels
        labels = list(set(encoder.transform(df[df.columns[index]])))
        # If nominal feature is binary -> encode with label encoder
        if (len(labels) == 2):
            dicts[df.columns[index]] = list(encoder.classes_)
            df[df.columns[index]] = encoder.transform(df[df.columns[index]])
            print('   Feature %s - encoded' % name)
        # If nominal feature is not binary -> one hot encoding
        else:
            # One hot encoding
            dummies = pd.get_dummies(df[df.columns[index]])
            # Rename headers
            headers = dummies.columns.tolist()
            for header in headers:
                if (name=='Mjob'):
                    headers[headers.index(header)] = 'MJ_' + header
                elif (name=='Fjob'):
                    headers[headers.index(header)] = 'FJ_' + header
                elif (name=='reason'):
                    headers[headers.index(header)] = header + '_reason'
                elif (name=='guardian'):
                    headers[headers.index(header)] = header + '_guardian'
            dummies.columns = headers
            # Join new features
            df = df.join(dummies)
            # Place new features after initial feature
            cols = df.columns.tolist()
            for header in headers[::-1]:
                cols.insert(index+1, cols.pop(cols.index(header)))
            # Reindex dataframe
            df = df.reindex(columns=cols)
            # Drop initial feature
            df = df.drop(df.columns[index], axis=1)
            print('   Feature %s - encoded' % name)
    print("Nominal features encoding completed. Total features encoded: ", len(column_names))
    return df, dicts
```

Повторим эксперимент из статьи. 

Введем целевой признак 'Alc'. Значение признака рассчитывается по формуле : Alc = (Walc x 2 + Dalc x 5)/7. 

Если результат меньше 3, то признак принимает значение 0, в других случаях признак принимает значение 1.

Также удалим признаки Walc и Dalc, так как от них напрямую зависит целевой признак (это можно заметить из формулы), а также признаки Fedu, G1, G2, G3, как это сделано в эксперименте, описанном в статье.

```python
# Adding target feature
df['alc'] = (df.Walc*2+df.Dalc*5)/7
df.alc[df.alc < 3] = 0
df.alc[df.alc >= 3] = 1

df = df.drop(['Dalc', 'Walc', 'Fedu', 'G1', 'G2', 'G3'], 1)
```

```python
# Split to train and test
RANDOM_SEED = 2016
X = df[df.columns[:-1]]
Y = df.alc
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=RANDOM_SEED)
```

Метод find_best_estimators находит 5 лучших настроек алгоритма по сетке и среди 5 находит самый лучший.

Метод compute_accuracies рассчитывает accuracy для 5 лучших настроек алгоритма.

Метод make_importances_table создает таблицу важностей признаков для модели и также строит график.

```python
def find_best_estimators(grid, estimator):
    best_clfs = []
    cv_scores = []
    for k in k_folds:
        cv = StratifiedKFold(Y_train, n_folds=k, shuffle=True, random_state=RANDOM_SEED)
        clf = GridSearchCV(estimator=estimator, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy')
        clf.fit(X_train, Y_train)
        # Find best classifier configuration  
        best_clfs.append(clf.best_estimator_)
        cv_scores.append(clf.best_score_)
    # Select best of the 5 best
    best_clf = best_clfs[np.argmax(cv_scores)]
    print('Наилучшие параметры:', best_clf)
    print('Лучшее качество при подборе параметров: ', np.max(cv_scores))
    return best_clf, best_clfs

def compute_accuracies(best_clfs, X_test, Y_test):
    accuracies = []
    for clf in best_clfs:
        accuracies.append(accuracy_score(Y_test, clf.predict(X_test)))
    print('Значение accuracy для пяти наилучших настроек метода на контрольной выборке:\n', accuracies)
    return accuracies

def make_importances_table(model, df):
    # Feature importances
    model_importances = model.feature_importances_
    # Making a plot
    idx = np.arange(model_importances.size)
    width = 0.8
    plt.figure(figsize=(19,8))
    p1 = plt.bar(idx, model_importances, width, color='g')
    plt.xticks(idx+width/2, np.arange(45))[1]
    
    # Create importances table
    importances_table = pd.DataFrame(data=df.columns[:-1], columns=['feature'])
    importances_table['importance'] = model_importances
    # Sort table in descending order
    importances_table = importances_table.sort('importance', ascending=False)
    for i in range(model_importances.size):
        importances_table.importance[i] = round(importances_table.importance[i]*100, 2)
        importances_table.importance[i] = str(importances_table.importance[i])+'%'
    return importances_table
```
